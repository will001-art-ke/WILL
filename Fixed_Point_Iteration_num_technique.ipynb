### Fixed-Point Iteration: 1D Case

A fixed-point iteration is a method for computing fixed points of a function. A fixed point of a function `g` is a value `x` such that `g(x) = x`. The iteration process is defined by the recurrence relation: `x_{n+1} = g(x_n)`.

We will use `g(x) = cos(x)` as an example. The fixed point of this function is approximately `0.739085`. Let's see how the iteration converges to this value.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function g(x)
def g(x):
    return np.cos(x)

# Initial guess
x0 = 0.3

# Number of iterations
num_iterations = 20

# Store the iteration values
iterations = [x0]

# Perform fixed-point iteration
for _ in range(num_iterations):
    x_next = g(iterations[-1])
    iterations.append(x_next)

print("1D Fixed-Point Iteration for g(x) = cos(x):")
for i, val in enumerate(iterations):
    print(f"Iteration {i}: x = {val:.6f}")

# Plotting the convergence
plt.figure(figsize=(8, 6))
plt.plot(range(len(iterations)), iterations, marker='o', linestyle='-', color='blue')
plt.title('Convergence of 1D Fixed-Point Iteration')
plt.xlabel('Iteration Number')
plt.ylabel('x value')
plt.grid(True)
plt.show()
```

### Fixed-Point Iteration: 2D Case and 3D Visualization

For a 2D system, we are looking for a point `(x, y)` such that `x = g_x(x, y)` and `y = g_y(x, y)`. The iteration becomes:

`x_{n+1} = g_x(x_n, y_n)`
`y_{n+1} = g_y(x_n, y_n)`

To provide a 3D visual of this technique, we'll track the `(x, y)` coordinates at each iteration step `n`. This creates a trajectory `(x_n, y_n, n)` in 3D space, showing how the iteration progresses towards a fixed point.

We'll use a simple system that converges:
`g_x(x, y) = 0.5 * np.cos(y) + 0.1 * x`
`g_y(x, y) = 0.5 * np.sin(x) + 0.1 * y`

This system is designed to contract and converge to a fixed point, typically near the origin.

```python
from mpl_toolkits.mplot3d import Axes3D

# Define the 2D fixed-point functions
def gx(x, y):
    return 0.5 * np.cos(y) + 0.1 * x

def gy(x, y):
    return 0.5 * np.sin(x) + 0.1 * y

# Initial guess for (x, y)
x0_2d, y0_2d = 0.8, 0.2

# Number of iterations
num_iterations_2d = 50

# Store iteration values for x, y, and iteration number
x_values = [x0_2d]
y_values = [y0_2d]
iteration_numbers = [0]

# Perform 2D fixed-point iteration
for i in range(num_iterations_2d):
    x_next = gx(x_values[-1], y_values[-1])
    y_next = gy(x_values[-1], y_values[-1])
    x_values.append(x_next)
    y_values.append(y_next)
    iteration_numbers.append(i + 1)

print("\n2D Fixed-Point Iteration Trajectory:")
for i in range(len(x_values)):
    print(f"Iteration {iteration_numbers[i]}: (x, y) = ({x_values[i]:.6f}, {y_values[i]:.6f})")

# Create the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the trajectory
ax.plot(x_values, y_values, iteration_numbers, marker='o', linestyle='-', color='red', alpha=0.7)

# Mark the start and end points
ax.scatter(x_values[0], y_values[0], iteration_numbers[0], color='green', s=100, label='Start Point (Iteration 0)')
ax.scatter(x_values[-1], y_values[-1], iteration_numbers[-1], color='purple', s=100, label='End Point (Converged)')

ax.set_xlabel('X Value')
ax.set_ylabel('Y Value')
ax.set_zlabel('Iteration Number')
ax.set_title('3D Visualization of 2D Fixed-Point Iteration Convergence')
ax.legend()
ax.grid(True)
plt.show()
```
