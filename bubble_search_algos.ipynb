```python
def linear_search(arr, target):
    """
    Implements linear search to find the target element in a list.

    Args:
        arr (list): The list to search within.
        target: The element to search for.

    Returns:
        int: The index of the target if found, otherwise -1.
    """
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

# Example usage:
my_list = list(range(10)) # Updated to 10 elements from 0 to 9
target_element = 7

index = linear_search(my_list, target_element)

if index != -1:
    print(f"Element {target_element} found at index {index}")
else:
    print(f"Element {target_element} not found in the list")

target_element_not_found = 10 # This element is not in the list 0-999
index_not_found = linear_search(my_list, target_element_not_found)

if index_not_found != -1:
    print(f"Element {target_element_not_found} found at index {index_not_found}")
else:
    print(f"Element {target_element_not_found} not found in the list")
```

```python
import random

def bubble_sort(arr):
    """
    Implements bubble sort algorithm with quadratic time complexity (O(n^2)).

    Args:
        arr (list): The list to be sorted.

    Returns:
        list: The sorted list.
    """
    n = len(arr)
    # Traverse through all array elements
    for i in range(n - 1):
        # Last i elements are already in place
        for j in range(0, n - i - 1):
            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# Example usage with 10 elements:
# Generate a list of 10 random integers for demonstration
list_of_100_elements = [random.randint(0, 10) for _ in range(10)]

print("Original list (first 10 elements):", list_of_100_elements[:10])

sorted_list = bubble_sort(list_of_100_elements)

print("Sorted list (first 10 elements):", sorted_list[:10])
print("Sorted list (last 10 elements):", sorted_list[-10:])
```
