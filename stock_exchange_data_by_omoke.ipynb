### Load Nairobi Stock Exchange Data for Safaricom (SCOM.NR) for 2024 using Polars

This cell now uses the `yfinance` library to download historical stock data and then converts it to a Polars DataFrame for efficient data handling.

```python
import yfinance as yf
import polars as pl

# Define the ticker symbol for a Nairobi Stock Exchange company
ticker_symbol = 'KCB.NR' # Changed from 'SCOM.NR' to 'KCB.NR'

# Define the start and end dates for 2024
start_date = '2024-01-01'
end_date = '2024-12-31' # Fetches data up to the current date if 2024 is ongoing

print(f"Fetching data for {ticker_symbol} from {start_date} to {end_date}...")

# Download the data
try:
    # yfinance.download returns a pandas DataFrame, convert it to polars
    pandas_data = yf.download(ticker_symbol, start=start_date, end=end_date)

    if not pandas_data.empty:
        data = pl.from_pandas(pandas_data)
        print(f"Successfully loaded data for {ticker_symbol} into a Polars DataFrame.")
        display(data.head())
        print("\nPolars DataFrame Schema:")
        display(data.schema)
    else:
        print(f"No data found for {ticker_symbol} in the specified date range.")
except Exception as e:
    print(f"An error occurred while fetching data: {e}")
    print("Please ensure the ticker symbol is correct and you have an internet connection.")
```

### Load London Stock Exchange Data for Vodafone (VOD.L) for 2024

This cell uses the `yfinance` library to download historical stock data for Vodafone, a major company listed on the London Stock Exchange.

- `VOD.L` is the Yahoo Finance ticker symbol for Vodafone on the London Stock Exchange.
- The `start` and `end` parameters specify the date range for the data.
- `display(data.head())` shows the first few rows of the downloaded data.

```python
import yfinance as yf
import pandas as pd

# Define the ticker symbol for Vodafone on the London Stock Exchange
ticker_symbol = 'VOD.L'

# Define the start and end dates for 2024
start_date = '2024-01-01'
end_date = '2024-12-31' # Fetches data up to the current date if 2024 is ongoing

print(f"Fetching data for {ticker_symbol} from {start_date} to {end_date}...")

# Download the data
try:
    data = yf.download(ticker_symbol, start=start_date, end=end_date)
    if not data.empty:
        print(f"Successfully loaded data for {ticker_symbol}.")
        display(data.head())
        display(data.info())
    else:
        print(f"No data found for {ticker_symbol} in the specified date range.")
except Exception as e:
    print(f"An error occurred while fetching data: {e}")
    print("Please ensure the ticker symbol is correct and you have an internet connection.")
```

```python
# Calculate and display summary statistics for the VOD.L dataset
display(data.describe())
```

```python
display(data.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Select the columns to scale (Open, High, Low, Close, Volume)
columns_to_scale = ['Open', 'High', 'Low', 'Close', 'Volume']

# Create a copy to avoid modifying the original DataFrame directly
data_scaled = data.copy()

# Apply the scaler to the selected columns
data_scaled[columns_to_scale] = scaler.fit_transform(data_scaled[columns_to_scale])

print("Scaled data (first 5 rows):")
display(data_scaled.head())
```

```python
# Define the target variable (y) and features (X)
y = data_scaled['Volume']
X = data_scaled.drop(columns=['Volume'])

print("Target variable (y) head:")
display(y.head())

print("\nFeatures (X) head:")
display(X.head())
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets (e.g., 80% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# LSTM expects input in the shape of (samples, timesteps, features)
# Currently, X_train and X_test are (samples, features)
# For a simple case, we can treat each sample as a single timestep.

timesteps = 1 # Each sample is considered a single timestep

X_train_reshaped = np.reshape(X_train, (X_train.shape[0], timesteps, X_train.shape[1]))
X_test_reshaped = np.reshape(X_test, (X_test.shape[0], timesteps, X_test.shape[1]))

print("Reshaped X_train shape:", X_train_reshaped.shape)
print("Reshaped X_test shape:", X_test_reshaped.shape)
```

Now, let's define the architecture of our LSTM model. We'll use a `Sequential` model with one LSTM layer and one Dense output layer. The output layer will predict the 'Volume'.

```python
# Define the LSTM model
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(timesteps, X_train.shape[1])))
model.add(Dense(1)) # Output layer for predicting 'Volume'

# Compile the model
model.compile(optimizer='adam', loss='mse')

model.summary()
```

Next, we will train the LSTM model using the `X_train_reshaped` and `y_train` data. We'll use a batch size of 32 and train for 50 epochs.

```python
# Train the model
history = model.fit(X_train_reshaped, y_train, epochs=100, batch_size=32, verbose=1, validation_split=0.1)

print("Model training complete.")
```

### Make Predictions and Evaluate the Model

```python
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Make predictions on the test data
y_pred_scaled = model.predict(X_test_reshaped)

# Evaluate the model on the scaled test data
mse = mean_squared_error(y_test, y_pred_scaled)
r2 = r2_score(y_test, y_pred_scaled)

print(f"Mean Squared Error on scaled test data: {mse:.4f}")
print(f"R-squared on scaled test data: {r2:.4f}")
```

```python
from sklearn.metrics import mean_absolute_error

# Calculate Mean Absolute Error on the scaled test data
mae = mean_absolute_error(y_test, y_pred_scaled)

print(f"Mean Squared Error on scaled test data: {mse:.4f}")
print(f"Mean Absolute Error on scaled test data: {mae:.4f}")
print(f"R-squared on scaled test data: {r2:.4f}")
```

#### Visualize Actual vs. Predicted Scaled Volume

```python
# Plot actual vs. predicted scaled values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_scaled, alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--') # Diagonal line
plt.xlabel('Actual Scaled Volume')
plt.ylabel('Predicted Scaled Volume')
plt.title('Actual vs. Predicted Scaled Volume on Test Set')
plt.grid(True)
plt.show()
```

### Predicting Next Day's Closing Price with a New LSTM Model

For this task, we will create sequences of historical data to predict a future value. We'll use a sliding window approach where the model looks at the past `n` days of 'Close' prices to predict the 'Close' price on day `n+1`.

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt

# Reset index to make 'Date' a regular column for easier slicing
data_reset = data.reset_index()

# Select the 'Close' price for forecasting
close_prices = data_reset['Close'].values.reshape(-1, 1)

# Scale the 'Close' prices
scaler_close = MinMaxScaler(feature_range=(0, 1))
scaled_close_prices = scaler_close.fit_transform(close_prices)

# Define the number of past days to use for prediction
n_steps = 30

# Function to create sequences for LSTM
def create_sequences(data, n_steps):
    X, y = [], []
    for i in range(len(data) - n_steps):
        X.append(data[i:(i + n_steps), 0])
        y.append(data[i + n_steps, 0])
    return np.array(X), np.array(y)

# Create sequences
X_seq, y_seq = create_sequences(scaled_close_prices, n_steps)

print(f"Shape of X (sequences): {X_seq.shape}")
print(f"Shape of y (target): {y_seq.shape}")

display(X_seq[:5])
display(y_seq[:5])
```

Now we will split the data into training and testing sets. For time series data, it's crucial to maintain the chronological order, so we'll use a sequential split rather than a random one. We'll use 80% for training and 20% for testing.

```python
# Split data into training and testing sets (chronological split)
train_size = int(len(X_seq) * 0.8)
X_train_seq, X_test_seq = X_seq[0:train_size], X_seq[train_size:len(X_seq)]
y_train_seq, y_test_seq = y_seq[0:train_size], y_seq[train_size:len(y_seq)]

print(f"X_train_seq shape: {X_train_seq.shape}")
print(f"X_test_seq shape: {X_test_seq.shape}")
print(f"y_train_seq shape: {y_train_seq.shape}")
print(f"y_test_seq shape: {y_test_seq.shape}")

# Reshape input to be [samples, timesteps, features] for LSTM
X_train_seq = X_train_seq.reshape(X_train_seq.shape[0], X_train_seq.shape[1], 1)
X_test_seq = X_test_seq.reshape(X_test_seq.shape[0], X_test_seq.shape[1], 1)

print(f"Reshaped X_train_seq shape: {X_train_seq.shape}")
print(f"Reshaped X_test_seq shape: {X_test_seq.shape}")
```

Next, we will define and compile our new LSTM model for predicting the next day's closing price. We will use a simple architecture with one LSTM layer and one Dense output layer.

```python
# Define the LSTM model for next day price prediction
model_close_price = Sequential()
model_close_price.add(LSTM(50, activation='relu', return_sequences=True, input_shape=(n_steps, 1)))
model_close_price.add(LSTM(50, activation='relu'))
model_close_price.add(Dense(1)) # Output layer for predicting the next day's 'Close' price

# Compile the model
model_close_price.compile(optimizer='adam', loss='mse')

model_close_price.summary()
```

Now, let's train the `model_close_price` using our prepared sequences.

```python
# Train the model
history_close_price = model_close_price.fit(X_train_seq, y_train_seq, epochs=100, batch_size=32, verbose=1, validation_split=0.1)

print("Model training complete.")
```

### Visualize Training History to Check for Overfitting

```python
import matplotlib.pyplot as plt

# Plot training & validation loss values
plt.figure(figsize=(10, 6))
plt.plot(history_close_price.history['loss'], label='Train Loss')
plt.plot(history_close_price.history['val_loss'], label='Validation Loss')
plt.title('Model Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss (Mean Squared Error)')
plt.legend()
plt.grid(True)
plt.show()
```

### Evaluate the Next Day Closing Price Prediction Model

Now we will evaluate the performance of our `model_close_price` on the test set and visualize the predictions against actual values.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Make predictions on the test data
y_pred_close_scaled = model_close_price.predict(X_test_seq)

# Evaluate the model on the scaled test data
mse_close = mean_squared_error(y_test_seq, y_pred_close_scaled)
mae_close = mean_absolute_error(y_test_seq, y_pred_close_scaled)
r2_close = r2_score(y_test_seq, y_pred_close_scaled)

print(f"Mean Squared Error on scaled test data (Close Price): {mse_close:.4f}")
print(f"Mean Absolute Error on scaled test data (Close Price): {mae_close:.4f}")
print(f"R-squared on scaled test data (Close Price): {r2_close:.4f}")

# Inverse transform the predictions and actual values to their original scale
y_test_close_unscaled = scaler_close.inverse_transform(y_test_seq.reshape(-1, 1))
y_pred_close_unscaled = scaler_close.inverse_transform(y_pred_close_scaled)

print("\nUnscaled y_test (first 5):\n", y_test_close_unscaled[:5].flatten())
print("Unscaled y_pred (first 5):\n", y_pred_close_unscaled[:5].flatten())
```

```python
# Plot actual vs. predicted unscaled closing prices
plt.figure(figsize=(12, 7))
plt.plot(y_test_close_unscaled, label='Actual Close Price')
plt.plot(y_pred_close_unscaled, label='Predicted Close Price')
plt.title('Actual vs. Predicted Next Day Close Price (Unscaled)')
plt.xlabel('Time Step (Test Data)')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()
```

### Hyperparameter Tuning: Learning Rate and Batch Size

Hyperparameter tuning is crucial for optimizing model performance. We'll focus on `learning_rate` and `batch_size`.

*   **Learning Rate**: Controls how much to change the model in response to the estimated error each time the model weights are updated. A smaller learning rate requires more training epochs but can lead to a more optimal set of weights. A larger learning rate may train faster but could overshoot the optimal solution.
*   **Batch Size**: The number of samples processed before the model is updated. Smaller batches introduce more noise but can help the model escape local minima. Larger batches provide a more accurate estimate of the gradient but can lead to poorer generalization.

We will define a function to create and compile the LSTM model, allowing us to easily experiment with different learning rates. Then, we'll perform a simple grid search over a few selected learning rates and batch sizes.

```python
from tensorflow.keras.optimizers import Adam

def create_tuned_lstm_model(learning_rate):
    """
    Creates and compiles an LSTM model with a specified learning rate.
    """
    model = Sequential()
    model.add(LSTM(50, activation='relu', return_sequences=True, input_shape=(n_steps, 1)))
    model.add(LSTM(50, activation='relu'))
    model.add(Dense(1))

    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='mse')
    return model

print("Model creation function defined.")
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Define hyperparameters to search
learning_rates_to_try = [0.01, 0.001, 0.0001]
batch_sizes_to_try = [16, 32, 64]

best_loss = float('inf')
best_hyperparameters = {}
best_model = None

results = []

for lr in learning_rates_to_try:
    for bs in batch_sizes_to_try:
        print(f"\n--- Training with Learning Rate: {lr}, Batch Size: {bs} ---")

        # Create a new model for each combination
        current_model = create_tuned_lstm_model(learning_rate=lr)

        # Define Early Stopping callback
        early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

        # Train the model
        history = current_model.fit(X_train_seq, y_train_seq,
                                    epochs=100, # Increased epochs but relying on EarlyStopping
                                    batch_size=bs,
                                    verbose=0, # Set to 0 to suppress output during tuning loop
                                    validation_split=0.1,
                                    callbacks=[early_stopping])

        # Evaluate on test set
        loss = current_model.evaluate(X_test_seq, y_test_seq, verbose=0)

        # Store results
        results.append({
            'learning_rate': lr,
            'batch_size': bs,
            'test_loss': loss,
            'epochs_trained': len(history.history['loss'])
        })

        print(f"Test Loss: {loss:.4f}")

        # Check if this is the best model
        if loss < best_loss:
            best_loss = loss
            best_hyperparameters = {'learning_rate': lr, 'batch_size': bs}
            best_model = current_model # Store the best model

print("\n--- Hyperparameter Tuning Complete ---")
print(f"Best Hyperparameters: {best_hyperparameters}")
print(f"Best Test Loss: {best_loss:.4f}")
```

After running the tuning process, `best_model` now holds the model trained with the optimal hyperparameters found during the search. You can now use `best_model` for further evaluation and prediction, or inspect the `results` list to see the performance of each hyperparameter combination.

### Evaluate the Best Hyperparameter Tuned Model

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Make predictions on the test data using the best model
y_pred_best_scaled = best_model.predict(X_test_seq)

# Evaluate the best model on the scaled test data
mse_best = mean_squared_error(y_test_seq, y_pred_best_scaled)
mae_best = mean_absolute_error(y_test_seq, y_pred_best_scaled)
r2_best = r2_score(y_test_seq, y_pred_best_scaled)

print(f"Mean Squared Error on scaled test data (Best Model - Close Price): {mse_best:.4f}")
print(f"Mean Absolute Error on scaled test data (Best Model - Close Price): {mae_best:.4f}")
print(f"R-squared on scaled test data (Best Model - Close Price): {r2_best:.4f}")

# Inverse transform the predictions and actual values to their original scale
y_test_close_unscaled = scaler_close.inverse_transform(y_test_seq.reshape(-1, 1))
y_pred_best_unscaled = scaler_close.inverse_transform(y_pred_best_scaled)

print("\nUnscaled y_test (first 5):\n", y_test_close_unscaled[:5].flatten())
print("Unscaled y_pred (Best Model - first 5):\n", y_pred_best_unscaled[:5].flatten())
```
