```python
def count_sort(arr):
    # If the input array is empty, there's nothing to sort
    if not arr:
        return []

    # Find the maximum element in the input array to determine the range
    max_val = max(arr)

    # Create a count array initialized with zeros. Its size will be max_val + 1
    # to accommodate numbers from 0 to max_val.
    count = [0] * (max_val + 1)

    # Populate the count array: count occurrences of each element
    for num in arr:
        count[num] += 1

    # Modify the count array to store the actual position of each element
    # in the output array. This makes it a running sum.
    for i in range(1, len(count)):
        count[i] += count[i-1]

    # Create an output array of the same size as the input array
    output = [0] * len(arr)

    # Iterate through the input array in reverse to maintain stability
    # (i.e., elements with the same value appear in the same order as in the input).
    for num in reversed(arr):
        # Get the position for the current number from the count array
        position = count[num] - 1
        # Place the number into the output array
        output[position] = num
        # Decrement the count for this number, as we've placed one instance
        count[num] -= 1

    return output

# The previous global 'arr' and 'ans' from the kernel state are still available.
# Let's clean up the old 'ans' variable if it's not needed for demonstration.
# The actual demonstration will be in a new cell.
```

```python
arr_to_sort = [4, 2, 2, 8, 3, 3, 1, 0, 5, 4]
sorted_arr = count_sort(arr_to_sort)
print(f"Original array: {arr_to_sort}")
print(f"Sorted array using count sort: {sorted_arr}")
```

```python
def countingSort_for_radix(arr, exp):
    n = len(arr)
    output = [0] * n
    count = [0] * 10  # For digits 0-9

    # Store count of occurrences in count[]
    for i in range(n):
        index = arr[i] // exp
        count[index % 10] += 1

    # Change count[i] so that count[i] now contains actual
    # position of this digit in output array
    for i in range(1, 10):
        count[i] += count[i - 1]

    # Build the output array
    i = n - 1
    while i >= 0:
        index = arr[i] // exp
        output[count[index % 10] - 1] = arr[i]
        count[index % 10] -= 1
        i -= 1

    # Copying the output array to arr[], so that arr now
    # contains sorted numbers according to current digit
    for i in range(n):
        arr[i] = output[i]

def radixSort(arr):
    # Find the maximum number to know number of digits
    max1 = max(arr)

    # Do counting sort for every digit. Note that instead
    # of passing digit number, exp is passed. exp is 10^i
    # where i is current digit number
    exp = 1
    while max1 // exp > 0:
        countingSort_for_radix(arr, exp)
        exp *= 10

# Example usage:
arr_radix = [170, 45, 75, 90, 802, 24, 2, 66]
radixSort(arr_radix)
print(f"Original array (for Radix Sort): [170, 45, 75, 90, 802, 24, 2, 66]")
print(f"Sorted array using Radix Sort: {arr_radix}")
```

```python
def bubble_sort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n - i - 1):
            # Traverse the array from 0 to n-i-1
            # Swap if the element found is greater
            # than the next element
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

# Example usage:
example_arr = [64, 34, 25, 12, 22, 11, 90]
print(f"Original array (for Bubble Sort): {example_arr}")
bubble_sort(example_arr)
print(f"Sorted array using Bubble Sort: {example_arr}")
```
